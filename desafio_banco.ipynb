```python
""""
    Desafio #1
    Criar um sistema de banco que permita dep√≥sitos, saques e consulta ao extrato
    O limite de saque √© de 500 por opera√ß√£o, e o m√°ximo de opera√ß√µes de saque √© de 3 por dia.

    Desafio #2
    Estruturar o c√≥digo atrav√©s de fun√ß√µes
    Criar o cadastro de usu√°rio e de contas
"""

def menu():
    menu = """
    [d] Depositar
    [s] Sacar
    [e] Extrato
    [q] Sair
    [conta] Nova Conta
    [lista] Listar Contas
    [novo] Novo Usu√°rio
    => """
    return input(menu)

def depositar(valor, saldo, extrato, /):
    saldo += valor
    extrato += f"Dep√≥sito: R$ {valor:.2f}\n"
    print(f"‚úÖ Dep√≥sito de R$ {valor:.2f} realizado com sucesso!")

    return saldo, extrato

def sacar(*, valor, saldo, extrato, limite, numero_saques, limite_saques):
    saldo -= valor
    extrato += f"Saque: R$ {valor:.2f}\n"
    numero_saques += 1
    print(f"‚úÖ Saque de R$ {valor:.2f} realizado com sucesso!")
    
    return saldo, extrato

def exibir_extrato(saldo, /, *, extrato):
    print("\n================ EXTRATO ================")
    print("üìå N√£o foram realizadas movimenta√ß√µes." if not extrato else extrato)
    print(f"\nüí∞ Saldo atual: R$ {saldo:.2f}")
    print("==========================================")

    return saldo, extrato

def criar_usuario(usuarios):
    
    cpf = input("Informe o CPF").strip()
    usuario = filtrar_usuarios(cpf, usuarios)

    if usuario:
        print("‚ö†Ô∏è J√° existe um usu√°rio com este CPF")
        input("Pressione Enter para continuar...")
        return

    nome = input("Digite seu nome").strip()
    nascimento = input("Informe a data de nascimento dd-mm-aaaa").strip()
    endereco = input("Informe o endere√ßo").strip()

    usuarios.append({"cpf": cpf, "nome": nome, "nascimento": nascimento, "endereco": endereco})

    print(f"‚úÖ Usu√°rio cadastrado com sucesso!")

def filtrar_usuarios(cpf, usuarios):
    cpf = cpf.strip()
    return next((usuario for usuario in usuarios if usuario["cpf"] == cpf), None)

def criar_conta(agencia, conta, usuarios):

    cpf = input("Informe o CPF").strip()
    usuario = filtrar_usuarios(cpf, usuarios)

    if usuario:
        print(f"‚úÖ Conta criada com sucesso!")
        input("Pressione Enter para continuar...")
        return {"agencia": agencia, "conta": conta, "usuario": usuario}
    
    print("‚ö†Ô∏è Usu√°rio n√£o encontrado. Verifique e tente novamente.")

def listar_contas(contas):
    for conta in contas:
        resultado = f"""\
            Ag√™ncia: {conta["agencia"]} 
            N¬∫ da conta: {conta["conta"]} 
            Usu√°rio: {conta["usuario"]["nome"]}"""

        print(resultado)

def main():

    saldo = 0
    limite = 500
    extrato = ""
    numero_saques = 0
    usuarios = []
    contas = []
    LIMITE_SAQUES = 3
    AGENCIA = "0001"

    while True:
        opcao = menu()

        if opcao == "d":
            while True:
                try:
                    valor = float(input("Informe o valor: "))
                    if valor > 0:
                        saldo, extrato = depositar(valor, saldo, extrato)
                        break
                    else:
                        print("‚ö†Ô∏è Informe um valor positivo para depositar.")
                except ValueError:
                    print("‚ö†Ô∏è Erro: Digite um n√∫mero v√°lido.")

        elif opcao == "s":
            while True:
                try:
                    valor = float(input("Informe o valor: "))
                    if numero_saques >= LIMITE_SAQUES:
                        print("‚ö†Ô∏è N√∫mero m√°ximo de saques atingido.")
                        break
                    elif valor <= 0:
                        print("‚ö†Ô∏è Informe um valor positivo para sacar.")
                    elif valor > saldo:
                        print("‚ö†Ô∏è Saldo insuficiente.")
                    elif valor > limite:
                        print("‚ö†Ô∏è Limite de saque excedido (m√°x: R$ 500,00).")
                    else:
                        saldo, extrato = sacar(
                            valor = valor, 
                            saldo = saldo, 
                            extrato = extrato,
                            limite = limite, 
                            numero_saques = numero_saques,
                            limite_saques = LIMITE_SAQUES)
                        break
                except ValueError:
                    print("‚ö†Ô∏è Erro: Digite um n√∫mero v√°lido.")

        elif opcao == "e":
            exibir_extrato(saldo, extrato=extrato)

        elif opcao == "novo":
            criar_usuario(usuarios)

        elif opcao == "conta":
            conta = len(contas) + 1
            conta = criar_conta(AGENCIA, conta, usuarios)

            if conta:
                contas.append(conta)

        elif opcao == "lista":
            listar_contas(contas)

        elif opcao == "q":
            print("üëã Saindo... Obrigado por utilizar nosso sistema!")
            break

        else:
            print("‚ö†Ô∏è Op√ß√£o inv√°lida! Escolha uma das op√ß√µes dispon√≠veis.")

main()
```
